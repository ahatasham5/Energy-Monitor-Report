```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "/content/Short-Notesrevised.csv"
```

```python
df = pd.read_csv(path)
df
```

```python
plt.figure(figsize=(15,6))
plt.hist(df['Workload'], weights=df['Watt'])
plt.xlabel('Workload')
plt.ylabel('Watt')
plt.title('Histogram of Watt by Workload')
plt.grid(True)
```
